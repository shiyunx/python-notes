```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data file
df = pd.read_csv("studentsperformance.csv")
df
```

```python
# Groupby
df_test = df[["gender", "math score", "reading score"]]
df_group = df_test.groupby(["gender", "math score"], as_index=False).mean()
df_group
```

```python
# Pivot table
df_pivot = df_group.pivot(index="gender", columns="math score")
df_pivot
```

```python
# Heatmap
plt.pcolor(df_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```
