```python
# Lower
text = "AVOCADO"
text = str.lower(text)
print(text)
```

```python
# Upper
text = "mango"
text = str.upper(text)
print(text)
```

```python
# Title
text = "mango"
text = str.title(text)
print(text)
```

```python
# Capitalize
text = "mango"
text = str.capitalize(text)
print(text)
```

```python
# Remove extra spaces
text = "mango yellow"
text = str.strip(text)
print(text)
```

```python
# If else
user_input = int(input("Enter a number between 10 and 20: "))
if user_input >= 1 and user_input <= 10:
  print("Correct")
else:
  print("Incorrect")
```

```python
# If elif else
user_input = int(input("Enter a number between 1 and 5: "))
if user_input > 5:
  print("input is more than 5")
elif user_input == 5:
  print("input is equal to 5")
else:
  print("input is less than 5")
```

```python
# For list
fruits = ["apple", "orange", "mango", "avocado"]
for x in fruits:
  print(x)
```

```python
# While
i = 1
while i < 5:
  print(i)
  i = i +1
```

```python
# While
while True:
  user_input = int(input("Enter a number between 1 and 10: "))
  if user_input >= 1 and user_input <= 10:
    print("Correct")
    break
  else:
    print("Incorrect")
```

```python
# Add
fruits = {
    1:"apple",
    2:"orange",
    3:"mango",
    4:"avocado"
}

fruits[5] = "durian"
print(fruits)
```
