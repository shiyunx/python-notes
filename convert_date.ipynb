**Convert date to date format**

`pd.to_datetime(column)`

```python
# Import library
import pandas as pd
```

```python
# Date in text string format
df = pd.DataFrame({
    "exam_date": ["2025-12-21"]
})
```

```python
# Read data
df
```

```python
# Before: check datatype
df.dtypes
```

```python
# Convert text to date
df["exam_date"] = pd.to_datetime(df["exam_date"])
```

```python
# After: check datatype
df.dtypes
```

```python
# Extract year
df["year"] = df["exam_date"].dt.year
df
```

```python
# Extract month
df["month"] = df["exam_date"].dt.month
df
```

```python
# Extract day
df["day"] = df["exam_date"].dt.day
df
```
